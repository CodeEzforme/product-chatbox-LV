```python
import os
import pandas as pd
import random

# ==== CẤU HÌNH ====
image_folder = "C:/HOCTAP/Back-end/mohinh/items/data/image/nem_chua"  # Thư mục ảnh chính (có nhiều thư mục con)
questions_file = "C:/HOCTAP/Back-end/mohinh/items/data/câu hỏi/nem chua/questions_answers_train_nem.csv"  # Thay bằng file chứa câu hỏi
output_folder = "C:/HOCTAP/Back-end/mohinh/items/data/input"  # Thư mục để lưu file CSV

# ==== KIỂM TRA THƯ MỤC ====
if not os.path.exists(image_folder):
    raise ValueError(f"❌ Thư mục ảnh không tồn tại: {image_folder}")

if not os.path.exists(questions_file):
    raise ValueError(f"❌ File câu hỏi không tồn tại: {questions_file}")

if not os.path.exists(output_folder):
    os.makedirs(output_folder)  # Tạo thư mục nếu chưa có

# ==== ĐỌC DANH SÁCH ẢNH ====
valid_extensions = ('.jpg', '.jpeg', '.png')
image_files = []

# Duyệt qua tất cả thư mục con để lấy ảnh
for subdir, _, files in os.walk(image_folder):
    for file in files:
        if file.lower().endswith(valid_extensions):  # Chỉ lấy file ảnh hợp lệ
            full_path = os.path.join(subdir, file)
            image_files.append(full_path)

if len(image_files) == 0:
    raise ValueError("❌ Không tìm thấy file ảnh hợp lệ!")

print(f"✅ Tìm thấy {len(image_files)} file ảnh hợp lệ!")

# ==== ĐỌC CÂU HỎI TỪ FILE CSV ====
df_questions = pd.read_csv(questions_file)

# Kiểm tra file CSV có các cột đúng không
required_columns = ["sản phẩm", "câu hỏi", "trả lời"]
missing_columns = [col for col in required_columns if col not in df_questions.columns]

if missing_columns:
    raise ValueError(f"❌ File CSV thiếu các cột: {missing_columns}")

# Lấy danh sách dữ liệu từ file câu hỏi
questions = df_questions[["sản phẩm", "câu hỏi", "trả lời"]].dropna().values.tolist()  # Giữ lại cột "sản phẩm"

if len(questions) == 0:
    raise ValueError("❌ Không có dữ liệu hợp lệ trong file CSV!")

print(f"✅ Đọc được {len(questions)} câu hỏi và câu trả lời!")

# ==== TẠO DỮ LIỆU ====
data = []
num_samples_per_image = 3  # Mỗi ảnh gán với 3 câu hỏi (có thể thay đổi)

for img_path in image_files:
    for _ in range(num_samples_per_image):
        product, question, answer = random.choice(questions)  # Lấy cả cột sản phẩm
        data.append([img_path, product, question, answer])

print(f"✅ Tạo {len(data)} mẫu dữ liệu!")

# ==== CHIA DỮ LIỆU TRAIN/VAL/TEST (KHÔNG XÁO TRỘN) ====
train_split = int(0.7 * len(data))  # 70% train
val_split = int(0.15 * len(data))   # 15% validation
test_split = len(data) - train_split - val_split  # 15% test

train_data = data[:train_split]
val_data = data[train_split:train_split + val_split]
test_data = data[train_split + val_split:]

# ==== LƯU FILE CSV ====
def save_csv(filename, dataset):
    df = pd.DataFrame(dataset, columns=["image_path", "sản phẩm", "question", "answer"])
    df.to_csv(os.path.join(output_folder, filename), index=False)

save_csv("nem_train_vqa_dataset.csv", train_data)
save_csv("nem_val_vqa_dataset.csv", val_data)
save_csv("nem_vqa_dataset.csv", test_data)

print(f"✅ File CSV đã được tạo và lưu tại: {output_folder}")
print(f"🔹 Train: {len(train_data)} mẫu")
print(f"🔹 Validation: {len(val_data)} mẫu")
print(f"🔹 Test: {len(test_data)} mẫu")
```

```python
import os
import torch
import torch.nn as nn
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import ViTModel, ViTImageProcessor, BertTokenizer, BertModel
from PIL import Image
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Kiểm tra xem có GPU hay không
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load ViT model and feature extractor
feature_extractor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model_vit = ViTModel.from_pretrained('google/vit-base-patch16-224').to(device)

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model_bert = BertModel.from_pretrained('bert-base-uncased').to(device)


# Define dataset class
class VQADataset(Dataset):
    def __init__(self, csv_file):
        self.data = pd.read_csv(csv_file)
        self.le = LabelEncoder()
        self.data['encoded_answer'] = self.le.fit_transform(self.data['answer'])  # Mã hóa câu trả lời

        # Kiểm tra nếu dataset có dữ liệu
        if len(self.data) == 0:
            raise ValueError("Dataset is empty. Check your CSV file.")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        # Load image and extract features using ViT
        image_path = row['image_path']

        try:
            image = Image.open(image_path).convert("RGB")
        except Exception as e:
            raise ValueError(f"Error loading image {image_path}: {e}")

        inputs = feature_extractor(images=image, return_tensors="pt").to(device)
        image_features = model_vit(**inputs).last_hidden_state[:, 0, :]  # Lấy đặc trưng đầu tiên

        # Tokenize question and extract features using BERT
        question = row['question']
        inputs = tokenizer(question, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
        question_features = model_bert(**inputs).last_hidden_state[:, 0, :]  # Lấy đặc trưng đầu tiên

        # Combine image and question features
        combined_features = torch.cat((image_features, question_features), dim=1).squeeze()

        # Encode label
        label = torch.tensor(self.data.iloc[idx]['encoded_answer'], dtype=torch.long).to(device)
        return combined_features, label

# Define VQA model class
class VQAModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(VQAModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)  # tăng từ 0,2 lên 0,5 để đánh giá
    
    def forward(self, combined_features):
        x = self.fc1(combined_features)
        x = self.relu(x)
        x = self.dropout(x) 
        x = self.fc2(x)
        return x  # Không cần sử dụng softmax ở đây vì CrossEntropyLoss đã xử lý


# Initialize datasets and dataloaders
train_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/train_vqa_dataset-test.csv')
val_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/val_vqa_dataset_test-fixed.csv')
test_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/test_vqa_dataset_test-fixed.csv')
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=8, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False)


# Initialize model
input_dim = 768 + 768  # 768 cho ViT và 768 cho BERT
hidden_dim = 512
output_dim = len(train_dataset.data['encoded_answer'].unique())  # Số lớp = số câu trả lời khác nhau
model_vqa = VQAModel(input_dim=input_dim, hidden_dim=hidden_dim, output_dim=output_dim).to(device)


# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_vqa.parameters(), lr=0.0005, weight_decay=1e-4) #giảm tốc độ học từ 0,001 xuống 0,0005 và thêm weight decay để tránh overfitting


# Early stopping parameters
patience = 15  # Number of epochs to wait for improvement
best_val_loss = float('inf')
epochs_without_improvement = 0

# Training loop with validation
num_epochs = 100
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    model_vqa.train()
    epoch_loss = 0.0
    for i, (features, labels) in enumerate(train_dataloader):
        features, labels = features.to(device), labels.to(device)
        optimizer.zero_grad()
        output = model_vqa(features)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    train_losses.append(epoch_loss / len(train_dataloader))

    # Validate after each epoch
    model_vqa.eval()
    val_loss = 0.0
    with torch.no_grad():
        for features, labels in val_dataloader:
            features, labels = features.to(device), labels.to(device) 
            outputs = model_vqa(features)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

    val_losses.append(val_loss / len(val_dataloader))
    print(f'Epoch [{epoch + 1}/{num_epochs}] -> Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}')


    # Save the model with epoch number in the file name
    model_save_path1 = f"model_epoch_{epoch + 1}.pth"
    torch.save({
        'epoch': epoch + 1,
        'model_state_dict': model_vqa.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': val_losses[-1],
    }, model_save_path1)
    print(f"Model saved to {model_save_path1}")

    # Early stopping check
    if val_losses[-1] < best_val_loss:
        best_val_loss = val_losses[-1]
        epochs_without_improvement = 0
    else:
        epochs_without_improvement += 1

    if epochs_without_improvement >= patience:
        print(f'Early stopping triggered at epoch {epoch + 1}')
        break

# Plot train and validation loss
plt.plot(range(1, num_epochs + 1), train_losses, label='Training Loss')
plt.plot(range(1, num_epochs + 1), val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss vs Epochs')
plt.legend()
plt.show()

# Evaluate model
def evaluate_model(model, dataloader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for features, labels in dataloader:
            features, labels = features.to(device), labels.to(device) 
            outputs = model(features)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%")

# Evaluate on test set
evaluate_model(model_vqa, test_dataloader)

# Save final model after training
model_save_path = '/kaggle/working/vqa_model_final.pth'
torch.save({
    'epoch': num_epochs,
    'model_state_dict': model_vqa.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'loss': val_losses[-1],
}, model_save_path)
print(f"Model saved to {model_save_path}")
```
