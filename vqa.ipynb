```python
import pandas as pd

# Äá»c file CSV (cáº­p nháº­t Ä‘Æ°á»ng dáº«n file cá»§a báº¡n)
csv_path = r"C:\HOCTAP\Back-end\mohinh\items\data\val_kaggle_sort_update_1.csv"
df = pd.read_csv(csv_path)

# Kiá»ƒm tra xem cÃ¡c cá»™t cáº§n thiáº¿t cÃ³ tá»“n táº¡i khÃ´ng
if 'question' not in df.columns or 'answer' not in df.columns:
    raise ValueError("Dataset cáº§n cÃ³ cÃ¡c cá»™t 'question' vÃ  'answer'.")

# Äáº¿m sá»‘ lÆ°á»£ng cÃ¢u há»i duy nháº¥t
unique_question_count = df['question'].nunique()
print(f"Tá»•ng sá»‘ cÃ¢u há»i khÃ¡c nhau trong táº­p dá»¯ liá»‡u: {unique_question_count}")

# Äáº¿m sá»‘ láº§n xuáº¥t hiá»‡n cá»§a má»—i cÃ¢u há»i
question_counts = df['question'].value_counts()

# Lá»c ra cÃ¡c cÃ¢u há»i xuáº¥t hiá»‡n nhiá»u hÆ¡n 1 láº§n
duplicate_questions = question_counts[question_counts > 1].index

# Táº¡o DataFrame má»›i chá»©a cÃ¢u há»i trÃ¹ng láº·p vÃ  táº¥t cáº£ cÃ¡c cÃ¢u tráº£ lá»i khÃ¡c nhau
duplicate_df = df[df['question'].isin(duplicate_questions)].groupby('question')['answer'].unique().reset_index()

# ThÃªm cá»™t sá»‘ lÆ°á»£ng cÃ¢u tráº£ lá»i khÃ¡c nhau
duplicate_df['num_unique_answers'] = duplicate_df['answer'].apply(len)

# LÆ°u káº¿t quáº£ ra file má»›i
output_path = "C:/HOCTAP/Back-end/mohinh/items/data/duplicate_questions_with_answers.csv"
duplicate_df.to_csv(output_path, index=False, encoding='utf-8')

print(f"File Ä‘Ã£ Ä‘Æ°á»£c lÆ°u táº¡i: {output_path}")
```

```python
import os
import torch
import torch.nn as nn
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import ViTModel, ViTImageProcessor, BertTokenizer, BertModel
from PIL import Image
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Kiá»ƒm tra xem cÃ³ GPU hay khÃ´ng
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load ViT model and feature extractor
feature_extractor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model_vit = ViTModel.from_pretrained('google/vit-base-patch16-224').to(device)

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model_bert = BertModel.from_pretrained('bert-base-uncased').to(device)


# Define dataset class
class VQADataset(Dataset):
    def __init__(self, csv_file):
        self.data = pd.read_csv(csv_file)
        self.le = LabelEncoder()
        self.data['encoded_answer'] = self.le.fit_transform(self.data['answer'])  # MÃ£ hÃ³a cÃ¢u tráº£ lá»i

        # Kiá»ƒm tra náº¿u dataset cÃ³ dá»¯ liá»‡u
        if len(self.data) == 0:
            raise ValueError("Dataset is empty. Check your CSV file.")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        # Load image and extract features using ViT
        image_path = row['image_path']

        try:
            image = Image.open(image_path).convert("RGB")
        except Exception as e:
            raise ValueError(f"Error loading image {image_path}: {e}")

        inputs = feature_extractor(images=image, return_tensors="pt").to(device)
        image_features = model_vit(**inputs).last_hidden_state[:, 0, :]  # Láº¥y Ä‘áº·c trÆ°ng Ä‘áº§u tiÃªn

        # Tokenize question and extract features using BERT
        question = row['question']
        inputs = tokenizer(question, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
        question_features = model_bert(**inputs).last_hidden_state[:, 0, :]  # Láº¥y Ä‘áº·c trÆ°ng Ä‘áº§u tiÃªn

        # Combine image and question features
        combined_features = torch.cat((image_features, question_features), dim=1).squeeze()

        # Encode label
        label = torch.tensor(self.data.iloc[idx]['encoded_answer'], dtype=torch.long).to(device)
        return combined_features, label

# Define VQA model class
class VQAModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(VQAModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)  # tÄƒng tá»« 0,2 lÃªn 0,5 Ä‘á»ƒ Ä‘Ã¡nh giÃ¡
    
    def forward(self, combined_features):
        x = self.fc1(combined_features)
        x = self.relu(x)
        x = self.dropout(x) 
        x = self.fc2(x)
        return x  # KhÃ´ng cáº§n sá»­ dá»¥ng softmax á»Ÿ Ä‘Ã¢y vÃ¬ CrossEntropyLoss Ä‘Ã£ xá»­ lÃ½


# Initialize datasets and dataloaders
train_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/train_vqa_dataset-test.csv')
val_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/val_vqa_dataset_test-fixed.csv')
test_dataset = VQADataset('C:/HOCTAP/Back-end/mohinh/items/data/test_vqa_dataset_test-fixed.csv')
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=8, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False)


# Initialize model
input_dim = 768 + 768  # 768 cho ViT vÃ  768 cho BERT
hidden_dim = 512
output_dim = len(train_dataset.data['encoded_answer'].unique())  # Sá»‘ lá»›p = sá»‘ cÃ¢u tráº£ lá»i khÃ¡c nhau
model_vqa = VQAModel(input_dim=input_dim, hidden_dim=hidden_dim, output_dim=output_dim).to(device)


# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_vqa.parameters(), lr=0.0005, weight_decay=1e-4) #giáº£m tá»‘c Ä‘á»™ há»c tá»« 0,001 xuá»‘ng 0,0005 vÃ  thÃªm weight decay Ä‘á»ƒ trÃ¡nh overfitting


# Early stopping parameters
patience = 15  # Number of epochs to wait for improvement
best_val_loss = float('inf')
epochs_without_improvement = 0

# Training loop with validation
num_epochs = 100
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    model_vqa.train()
    epoch_loss = 0.0
    for i, (features, labels) in enumerate(train_dataloader):
        features, labels = features.to(device), labels.to(device)
        optimizer.zero_grad()
        output = model_vqa(features)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    train_losses.append(epoch_loss / len(train_dataloader))

    # Validate after each epoch
    model_vqa.eval()
    val_loss = 0.0
    with torch.no_grad():
        for features, labels in val_dataloader:
            features, labels = features.to(device), labels.to(device) 
            outputs = model_vqa(features)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

    val_losses.append(val_loss / len(val_dataloader))
    print(f'Epoch [{epoch + 1}/{num_epochs}] -> Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}')


    # Save the model with epoch number in the file name
    model_save_path1 = f"model_epoch_{epoch + 1}.pth"
    torch.save({
        'epoch': epoch + 1,
        'model_state_dict': model_vqa.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': val_losses[-1],
    }, model_save_path1)
    print(f"Model saved to {model_save_path1}")

    # Early stopping check
    if val_losses[-1] < best_val_loss:
        best_val_loss = val_losses[-1]
        epochs_without_improvement = 0
    else:
        epochs_without_improvement += 1

    if epochs_without_improvement >= patience:
        print(f'Early stopping triggered at epoch {epoch + 1}')
        break

# Plot train and validation loss
plt.plot(range(1, num_epochs + 1), train_losses, label='Training Loss')
plt.plot(range(1, num_epochs + 1), val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss vs Epochs')
plt.legend()
plt.show()

# Evaluate model
def evaluate_model(model, dataloader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for features, labels in dataloader:
            features, labels = features.to(device), labels.to(device) 
            outputs = model(features)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%")

# Evaluate on test set
evaluate_model(model_vqa, test_dataloader)

# Save final model after training
model_save_path = '/kaggle/working/vqa_model_final.pth'
torch.save({
    'epoch': num_epochs,
    'model_state_dict': model_vqa.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'loss': val_losses[-1],
}, model_save_path)
print(f"Model saved to {model_save_path}")
```
